# PyCity Schools Analysis

- Analysis

​Charter schools demonstrated superior performance compared to district schools in both average math and reading scores, as well as in the percentage of students passing these subjects.​ 

Further investigation is needed to identify the factors influencing this performance. 

In addition, small and medium-sized schools (those with 2,000 or fewer students) consistently outperformed larger schools across all assessed areas. 

Interestingly, students attending schools with higher budgets (exceeding $630.00 per student) did not perform as well as those in schools with smaller budgets.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passMathStudents = school_data_complete[school_data_complete["math_score"] >= 70]
passMathStudents.head()
#passMathStudents.head()
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
# (number of students who passed math / total number of students) * 100
percentagePassedMath = (numPassMath / studentCount) * 100
percentagePassedMath
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passReadingStudents = school_data_complete[school_data_complete["reading_score"] >= 70]
passReadingStudents.head()
#passReadingStudents.head()
numPassReading = len(passReadingStudents)
numPassReading
```

```python
# Use the following to calculate the percentage of students that passed reading (hint: Look at how the math percentage was calculated)
# (number of students who passed reading / total number of students) * 100
percentagePassedReading = (numPassReading / studentCount) * 100
percentagePassedReading
```

```python
# use conditional to find all students who had a score of 70 or higher in math and reading
passingStudents = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
#passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# (number of students who passed math and reading / total number of students) * 100
percentagePassed = (numPassingStudents / studentCount) * 100
percentagePassed
```

```python
# display the per_school summary dataframe
per_school_summary.head()
```

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary for later aggregations
schoolSpendingDF = per_school_summary.copy()
schoolSpendingDF
```

```python
# Use "pd.cut" on the per_school_capita Series from ealier to categorize per student spending based on the bins
schoolSpendingDF["Spending Ranges (Per Student)"] = pd.cut(perSchoolCapita, spending_bins, labels=labels)
schoolSpendingDF
```

```python
# since mapping was used earlier, convert all of teh following formatted columns back to floats
schoolSpendingDF["Average Math Score"] = schoolSpendingDF["Average Math Score"].astype(float)
schoolSpendingDF["Average Reading Score"] = schoolSpendingDF["Average Reading Score"].astype(float)
schoolSpendingDF["% Passing Math"] = schoolSpendingDF["% Passing Math"].str.replace("%","").astype(float)
schoolSpendingDF["% Passing Reading"] = schoolSpendingDF["% Passing Reading"].str.replace("%","").astype(float)
schoolSpendingDF["% Overall Pssing"] = schoolSpendingDF["% Overall Passing"].str.replace("%", "").astype(float)
schoolSpendingDF
```

```python
spendingGroupedDF = schoolSpendingDF.groupby("Spending Ranges (Per Student)",observed=False).mean(numeric_only=True)
spendingGroupedDF
```

```python
# grab each series
spendingMathSeries = spendingGroupedDF["Average Math Score"]
spendingReadingSeries = spendingGroupedDF["Average Reading Score"]
spendingPassMathSeries = spendingGroupedDF["% Passing Math"]
spendingPassReadingSeries = spendingGroupedDF["% Passing Reading"]
spendingGroupedDF["% Overall Passing"] = (spendingGroupedDF["% Passing Math"] + spendingGroupedDF["% Passing Reading"]) / 2
```

```python
# Assemble into DataFrame
spendingSummaryDF = pd.DataFrame(
    {
        "Average Math Score": spendingMathSeries,
        "Average Reading Score": spendingReadingSeries,
        "% Passing Math": spendingPassMathSeries,
        "% Passing Reading": spendingPassReadingSeries,
        "% Overall Passing": (spendingPassMathSeries + spendingPassReadingSeries) / 2
    }
)

spendingSummaryDF["Average Math Score"] = spendingSummaryDF["Average Math Score"].map("{:.2f}".format)
spendingSummaryDF["Average Reading Score"] = spendingSummaryDF["Average Reading Score"].map("{:.2f}".format)
spendingSummaryDF["% Passing Math"] = spendingSummaryDF["% Passing Math"].map("{:.2f}%".format)
spendingSummaryDF["% Passing Reading"] = spendingSummaryDF["% Passing Reading"].map("{:.2f}%".format)
spendingSummaryDF["% Overall Passing"] = spendingSummaryDF["% Overall Passing"].map("{:.2f}%".format)

# Display results
spendingSummaryDF
```

## Scores by School Type

```python
# group by type of school
schoo_type = school_data_complete.groupby("type")

#calculations 
average_math_score = schoo_type['math_score'].mean()
average_reading_score = schoo_type['math_score'].mean()
pass_math_percent = school_data_complete[school_data_complete['math_score'] >= 70].groupby('type')['Student ID'].count()/schoo_type['Student ID'].count()*100
pass_read_percent = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('type')['Student ID'].count()/schoo_type['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('type')['Student ID'].count()/schoo_type['Student ID'].count()*100

# df build            
scores_schoo_type = pd.DataFrame({
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_math_percent,
    '% Passing Reading': pass_read_percent,
    "% Overall Passing": overall})
    
#reorder columns
scores_schoo_type = scores_schoo_type[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "% Overall Passing"
]]
scores_schoo_type.index.name = "Type of School"


#formating
scores_schoo_type.style.format({'Average Math Score': '{:.6f}', 
                              'Average Reading Score': '{:.6f}', 
                              '% Passing Math': '{:.6f}', 
                              '% Passing Reading':'{:.6f}', 
                              '% Overall Passing': '{:.6f}'})
```
